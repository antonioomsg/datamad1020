### 2. All the companies that have more than 5000 employees. Limit the search to 20 companies and sort them by **number of employees**.

```python
filter_q = {"number_of_employees":{"$gt":5000}}
cur_2 = db.companies.find(filter_q).sort([("number_of_employees",1)]).limit(20)
len(list(cur_2))
```

### 3. All the companies founded between 2000 and 2005, both years included. Retrieve only the `name` and `founded_year` fields.

```python
filter_q = {"founded_year":{"$gte":1971, "$lte":2005}}
project = {"name":1,"founded_year":1}
cur_3 = db.companies.find(filter_q,project)
#len(list(cur_3))
```

### 4. All the companies that had a Valuation Amount of more than 100.000.000 and have been founded before 2010. Retrieve only the `name` and `ipo` fields.

```python
filter_q = {"$and":[
    {"ipo.valuation_amount":{"$gt":100000000}},
    {"founded_year":{"$lt":2010}}
    ]}
project = {"name":1,"ipo.valuation_amount":1}
cur_3 = db.companies.find(filter_q,project).limit(1)
list(cur_3)
```

### 5. All the companies that have less than 1000 employees and have been founded before 2005. Order them by the number of employees and limit the search to 10 companies.

```python
filter= {"$and":[
    {"number_of_employees":{"$lt":1000}},
    {"founded_year":{"$lt":2005}}   
]}
cur_5 = db.companies.find(filter).sort([("number_of_employees",1)]).limit(10)
len(list(cur_5))
```

### 6. All the companies that don't include the `partners` field.

```python
cur_6= db.companies.find({"partners":{"$exists": "false"}})
len(list(cur_6))
```

### 7. All the companies that have a null type of value on the `category_code` field.

### 19. All the companies that have been founded between 2000 and 2010, but have not been acquired before 2011.

```python
cur_19=db.companies.find( {"$and":[
    {"founded_year":{"$gte":2000}},
    {"founded_year":{"$lte":2010}},
    {"acquisition.acquired_year":{"$gte":2011}}   
]})
len(list(cur_19))
```

### 20. All the companies that have been 'deadpooled' after the third year.

```python
filter_20 = {"$where":"this.deadpooled_year - this.founded_year>3"}
project={"name":1,"deadpooled_year":1,"founded_year":1}
cur_20= db.companies.find(filter_20,project).limit(5)

list(cur_20)
```
